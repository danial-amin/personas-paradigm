Creating the Data into a String format



```python
# ── Block 1: Data preparation ─────────────────────────────────────────────────
import pandas as pd
import json

# Load your dataset (replace with your actual file)
df = pd.read_csv("/content/HCI_2025_DigitalLifeSurvey_public.csv")

# --- Helper functions ---
def dataframe_to_sample(df, max_rows=1000):
    """Return a readable string of the first N rows."""
    return df.head(max_rows).to_string(index=False)

def aggregate_stats(df):
    """Summary statistics so the LLM understands distributions."""
    stats = {}
    for col in df.select_dtypes(include='number').columns:
        stats[col] = {
            "mean": round(df[col].mean(), 2),
            "min":  df[col].min(),
            "max":  df[col].max()
        }
    for col in df.select_dtypes(include='object').columns:
        stats[col] = df[col].value_counts().head(5).to_dict()
    return json.dumps(stats, indent=2)

# Build the data payload (reused in Block 2)
data_sample = dataframe_to_sample(df)
data_stats  = aggregate_stats(df)
total_rows  = len(df)

print(f"✅ Dataset ready: {total_rows} rows, {len(df.columns)} columns")
print(df.head(3))
```

## Generating the Personas




### Goal Driven Personas

```python
# ── Block 2: Load prompt and generate personas (JSON output) ──────────────────
import anthropic
import json
from google.colab import userdata

# --- Config ---
PRODUCT     = "a company offering digital life coaching services"
MODEL       = "claude-sonnet-4-6"
PROMPT_FILE = "/content/goal-driven-persona.txt"

# --- Client ---
client = anthropic.Anthropic(api_key="")

# --- Load prompt template ---
with open(PROMPT_FILE, "r") as f:
    prompt_template = f.read()

system_prompt = prompt_template.replace("[PRODUCT]", PRODUCT)

# --- Build user message ---
user_message = f"""Here is the dataset ({total_rows} rows total):

--- DATA SAMPLE ---
{data_sample}

--- SUMMARY STATISTICS ---
{data_stats}

Using the data above, generate the personas now.
"""

# --- Call the API ---
response = client.messages.create(
    model=MODEL,
    max_tokens=4096,
    system=system_prompt,
    messages=[
        {"role": "user", "content": user_message}
    ]
)

personas_output = response.content[0].text

with open("personas_output_goal_oriented.txt", "w") as f:
    f.write(personas_output)

print("Saved to personas_output_goal_oriented.txt ✓")
```
